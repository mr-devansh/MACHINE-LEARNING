```python
# MAKING FIGURES WITH NUMPY ARRAYS
# A. line plot
import numpy as np
# 1. create data
x = np.linspace(0,10,100)
print(x[:10])
# 2. plot data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
#B. Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# C. Bar plot
prices = {"almond":10, "peanut":8,"cashew":18}
fig, ax = plt.subplots()
ax.bar(prices.keys(), prices.values())
ax.set(title="Price List", ylabel="Price", xlabel="Items");
```

```python
# C. Bar plot
prices = {"almond":10, "peanut":8,"cashew":18}
fig, ax = plt.subplots()
ax.barh(list(prices.keys()), list(prices.values()))
ax.set(title="Price List", ylabel="Items", xlabel="Price");
```

```python
# D. Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x, bins=20);
```

```python
# E. Subplots option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(prices.keys(), prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# E. Subplots option 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(prices.keys(), prices.values())
ax[1,1].hist(np.random.randn(1000));
```

```python
# another dataset
heart_disease = pd.read_csv("../DATA/heart-disease.csv")
heart_disease.head()
```

```python
# histogram for distribution of age 
heart_disease["age"].plot.hist(bins=30)
```

```python
heart_disease.plot.hist(figsize=(5,30), subplots=True, bins=20);
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# pyplot method (plotting directly from the pandas dataframe)
over_50.plot(kind="scatter", x = "age", y="chol", c="target");
```

```python
# matplotlib method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
```

```python
# OO matplotlib method
fig, ax = plt.subplots(figsize=(10,6))
# plot
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# customize
ax.set(title="Heart disease and cholestrol level",
      xlabel="Age",
      ylabel="Cholestrol");
# legend
ax.legend(*scatter.legend_elements(), title="Target");
# horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# OO matplotlib method full power
# using thalach and chol in one axis and age on other
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True)
# plot ax0 data
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# customize
ax0.set(title="Cholestrol level",
      ylabel="Cholestrol");
# legend
ax0.legend(*scatter.legend_elements(), 
           title="Target");
# horizontal line
ax0.axhline(over_50["chol"].mean(),
            linestyle="--",
            color="black");

# plot ax1 data
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
ax1.set(title="Thalach level",
       xlabel="Age",
       ylabel="Thalach")
ax1.legend(*scatter.legend_elements(), 
           title="Target")
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--",
            color="black");
fig.suptitle("Heart Disease Analysis",
            fontsize=16,
            fontweight="bold");
```
